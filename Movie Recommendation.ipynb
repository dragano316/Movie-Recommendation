```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
movie_data=pd.read_csv('https://raw.githubusercontent.com/codeheroku/Introduction-to-Machine-Learning/master/Building%20a%20Movie%20Recommendation%20Engine/movie_dataset.csv')
```

```python
movie_data.head(5)
```

```python
movie_data.isnull().sum()
```

```python
features=['keywords','cast','genres','director']
for feature in features:
    movie_data[feature]=movie_data[feature].fillna(' ')
```

```python
def combine_features(row):
    print( row['director']+' '+row['keywords']+' '+row['genres']+' '+row['cast'])
    return row['director']+' '+row['keywords']+' '+row['genres']+' '+row['cast']
movie_data['combined_features']=movie_data.apply(combine_features,axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv=CountVectorizer()
```

```python
text=cv.fit_transform(movie_data['combined_features'])
```

```python
text.toarray()
```

```python
similarity_scores=cosine_similarity(text)
```

```python
similarity_scores
```

```python
def title(index):
    return movie_data[movie_data.index==index]["original_title"].values[0]
    
def index(title):
    return movie_data[movie_data.title==title]["index"].values[0]

movie_user_likes="Avatar"
```

```python
movie_index=index(movie_user_likes)
movie_index
```

```python
similar_movies=list(enumerate(similarity_scores[movie_index]))
```

```python
sorted_similar_movies=sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

```python
sorted_similar_movies
```

```python
i=0
for element in sorted_similar_movies:
    print(title(element[0]))
    i+=1
    if(i>50):
        break
```
